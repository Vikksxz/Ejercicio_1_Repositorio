```python
## 1) Un concecionario quiere hacer cuentas sobre lo vendido en el último mes.
##     Pero quiere saber la cantidad total sin IVA, la cantidad total con IVA y el IVA
##     Haz un programa en python que pregunte al usario cuantos coches ha vendido de cada modelo de la tabla
##     Coches - Precio e imprima por pantalla el total con IVA, el total sin IVA y el IVA
## Nota: Si no se ve correctamente la celda de arriba pulsa en el botón de edit de la celda.
#Ford Focus	            13000	            21%
#Ford Turneo	        29000	            21%
#Ford Capri Eléctrico	48000	            12%
#Ford Mustang	        57000	            35%

import math

num_coches = int(input('Introduce el número de coches que has vendido: '))
modelo_coche = input('Introduce el modelo del coche (Focus, Turneo, Capri, Mustang): ').strip().capitalize()

precios = {
    "Focus": 13000,
    "Turneo": 29000,
    "Capri": 48000,
    "Mustang": 57000
}

IVA_n = 0.21
IVA_e = 0.12
IVA_s = 0.035

if modelo_coche not in precios:
    print("Modelo de coche no válido.")
else:  
    precio_sin_IVA = num_coches * precios[modelo_coche]
    if modelo_coche in ["Focus", "Turneo"]:
        iva = IVA_n
    elif modelo_coche == "Capri":
        iva = IVA_e
    else:
        iva = IVA_s

    precio_con_IVA = precio_sin_IVA * (1 + iva)
    print(f'El precio sin IVA sería de {precio_sin_IVA} y el precio con IVA sería de {precio_con_IVA:.2f}')

```

```python
## 4)El colegio electoral estadounidense tiene un total de 538 compromisarios encargados de elegir
## al nuevo presidente. Los compromisarios se reparten de manera propocional a la población pero teniendo en cuenta
## que antes de empezar cada estado ya tendrá 2 asignados y que el Distrito de Columbia siempre tendrá 3.
## Escribe un programa que imprima el número de compromisarios por estado.
## Comprueba que te ha salido bien imprimiendo el número de compromisarios resultantes de tu programa.

## El primer array contiene la lista de estados y el segundo contiene la población correspondiente de cada uno
## en millones de personas.

## ¡Ojo! ¡Los compromisarios son personas!

estados_eeuu = [
    "Alabama", "Alaska", "Arizona", "Arkansas", "California", "Carolina del Norte",
    "Carolina del Sur", "Colorado", "Connecticut", "Dakota del Norte", "Dakota del Sur",
    "Delaware", "Florida", "Georgia", "Hawái", "Idaho", "Illinois", "Indiana", "Iowa",
    "Kansas", "Kentucky", "Luisiana", "Maine", "Maryland", "Massachusetts", "Míchigan",
    "Minnesota", "Misisipi", "Misuri", "Montana", "Nebraska", "Nevada", "Nueva Jersey",
    "Nueva York", "Nuevo Hampshire", "Nuevo México", "Ohio", "Oklahoma", "Oregón",
    "Pensilvania", "Rhode Island", "Tennessee", "Texas", "Utah", "Vermont", "Virginia",
    "Virginia Occidental", "Washington", "Wisconsin", "Wyoming", "Distrito de Columbia"
]

poblacion_estados = [
    5.03, 0.73, 7.28, 3.03, 39.51, 10.49, 5.28, 5.81, 3.61, 0.77,
    0.89, 1.00, 21.48, 10.62, 1.46, 1.79, 12.67, 6.73, 3.19, 2.93,
    4.51, 4.62, 1.37, 6.05, 7.02, 9.98, 5.71, 2.96, 6.15, 3.66,
    1.07, 1.96, 3.15, 8.88, 19.45, 1.38, 2.11, 11.69, 3.99, 4.24,
    12.8, 1.1, 6.83, 29.36, 3.21, 0.64, 8.64, 1.77, 7.88, 5.85,
    0.58, 0.71 
]


total_compromisarios = 538
compromisarios_minimos = 2
compromisarios_dc = 3
compromisarios_iniciales = compromisarios_minimos * len(estados_eeuu) + (compromisarios_dc - compromisarios_minimos)

compromisarios_restantes = total_compromisarios - compromisarios_iniciales

poblacion_total = sum(poblacion_estados) - poblacion_estados[-1]

compromisarios_por_estado = {}
for i, estado in enumerate(estados_eeuu):
    if estado == "Distrito de Columbia":
        compromisarios_por_estado[estado] = compromisarios_dc
    else:
        compromisarios_proporcionales = int((poblacion_estados[i] / poblacion_total) * compromisarios_restantes)
        compromisarios_por_estado[estado] = compromisarios_minimos + compromisarios_proporcionales

while sum(compromisarios_por_estado.values()) != total_compromisarios:
    if sum(compromisarios_por_estado.values()) > total_compromisarios:
        for estado in compromisarios_por_estado:
            if compromisarios_por_estado[estado] > compromisarios_minimos and estado != "Distrito de Columbia":
                compromisarios_por_estado[estado] -= 1
                break
    else:
        for estado in compromisarios_por_estado:
            if estado != "Distrito de Columbia":
                compromisarios_por_estado[estado] += 1
                break

for estado, compromisarios in compromisarios_por_estado.items():
    print(f"{estado}: {compromisarios} compromisarios")

print(f"Total de compromisarios: {sum(compromisarios_por_estado.values())}")






```
